# Task 1 : Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Task 2 : Load the dataset

```python
df = pd.read_csv('titanic.csv')
df
```

# Task 3 : Data Cleaning

- 3.1 Count the number of missing values

```python
df.isnull().sum()
```

- 3.2 Count the number of duplicate values

```python
print(df.duplicated().sum()) # Here we cam see there are no duplicate values in the dataset
```

- 3.3 Removing rows with missing values

```python
df = df.dropna()
df
```

- 3.4 Checking for missing values after cleaning

```python
df.isnull().sum()

# now we can see there are no null values left in the dataset
```

- 3.5 Removing column for better analysis

```python
df = df.drop(['Parch'], axis=1) 
df.head()
```

- 3.5 Display the cleaned dataset's first few rows

```python
print("Cleaned dataset (first few rows) : ")
df.head()
```

# Task 4 : Data Manipulation

- 4.1 Finding out the number of survivors and non- survivors

```python
df["Survived"].value_counts()  # Here 0 means dead and 1 means alive
```

- 4.2 Finding the total number of survivors 

```python
survive = df['Survived'] == 1
print(survive.value_counts().values[1])
```

- 4.3 Displaying the details of the survivors

```python
df[survive] # passengers that survived
```

# Task 5 : Descriptive Analysis

- 5.1 Descriptive Statistics for numerical  columns

```python
df.describe()
```

- 5.2 Counting the number of passengers in first class, second class and third class

```python
pclass = df['Pclass'].value_counts()
print(pclass)
```

# Task 6 : Data Visualization

- 6.1 Pie Chart for the Number of Passengers by Classes of Service

The pie chart will display the number of passengers categorized by classes. This visualization shows proportion of passengers of each class.

```python
plt.figure(figsize=(10,5))
pclass = df['Pclass'].value_counts()
label = ['first class','second class','third class']
plt.pie(pclass, autopct= '%.1f%%', labels=label)
plt.title('Passengers by Classes')
plt.show()
```

- 6.2 Scatter Plot of Fare vs. Age

The scatter plot will illustrate the relationship between the passenger fare and passenger age. 

```python
plt.figure(figsize=(12,5))
x = df['Age']
y = df['Fare']
plt.scatter(x, y)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Fare vs. Age')
plt.show()
```

- 6.3 Histogram for Passengers by Age

The histogram will visually represent the age of passengers.

```python
plt.figure(figsize=(12,5))
x = df['Age']
plt.hist(x, color='lightgreen', edgecolor='black')
plt.title('Passengers by Age')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```
